### seaborn 라이브러리
- 파이썬의 데이터 시각화 라이브러리 특히 통계적 그래프를 수비게 그릴수있다
- matplotlib 을 기반으로 하지만 더 간편한 시각화,방법이 간편 

```python
# 성별 -> 학력 -> 월급 
gender=a.groupby(['성별','학력'])['월급'].mean().reset_index()
```

```python
gender.unstack()
```

```python
sns.barplot(data=gender,x='학력',y='월급',hue='성별')
```

```python
data5=a.groupby(['나이','국적'])['월급'].mean().reset_index()
```

```python
sns.barplot(data=data5,x='월급',y='국적',hue='나이',palette="pastel")
plt.figure(figsize=(20,5))
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=a,x='연차',y='월급',hue='국적')
# 선뒤에 있는 음영부분은 신뢰구간을 나타낸다 
# 신뢰구간은 각선에 대한 데이터의 변동성이나 불확실성은을 나타낸다-> 월급이 특정연차에서 얼마나 일관성있는 데이터인지 보여준다 
```
