
np.unique()->   


```python
a=np.unique(b[:,0])
print(a)
```

```python
star=[]

for i in a:
    tmp=b[b[:,0]==i, 2].mean().round(2)
    temp=[i,tmp]
    star.append(temp)
```

```python
star=np.array(star).reshape(-1,2)
print(star)
```

```python
for i in range (len(star)):
    print(star[i])
```

```python
star=np.array(star).reshape(-1,2)
print(star)
```

```python
movie_code=np.unique(b[:,1])
movie_code_star=[]

for i in movie_code:
    tmp=b[b[:,1]==i, 2].mean().round(2)
    temp=[i,tmp]
    movie_code_star.append(temp)
```

```python
movie_code_star=np.array(movie_code_star).reshape(-1,2)
print(movie_code_star)
```
