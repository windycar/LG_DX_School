#### 간단 실습 
- 교육수준의 따른 평균연봉 

```python
b=a["EDU_LEVEL"].unique()
b
```

```python
a[a["EDU_LEVEL"]=='박사 학위']['SALARY']
```

```python
a[a["EDU_LEVEL"]=='박사 학위']['SALARY'].mean()
```

```python
for i in b:
    tmp=0
    tmp=(a[a["EDU_LEVEL"]==i]['SALARY'].mean()).round(2)
    print(f"{i} : {tmp}")
```

```python
a[["EDU_LEVEL",'SALARY']].groupby('EDU_LEVEL')['SALARY'].mean().round(2)
```

```python
data=a[["EDU_LEVEL",'SALARY','GENDER']].groupby(['EDU_LEVEL','GENDER'])['SALARY'].mean().round(2)
```

```python
data
```

```python
#성별에 따른 평균연봉
data=a[['SALARY','GENDER']].groupby('GENDER')['SALARY'].mean().round(2)
data
```

```python
#성별에 따른 연봉의 다양한 통계량 구하기 
data=a[['SALARY','GENDER']].groupby('GENDER')['SALARY'].agg(['mean','max','min','median'])
```

```python
data
```

```python
# 상관계수 확인 -> 데이터.corr()
# 선형정 관걔 확인
a[['PERFORMANCE_SCORE','SALARY']].corr()
```

```python
a.corr(numeric_only= True)
#절대값 1에 가까울수록 상관관계가 높다 
#근속연수가 높을수록 연봉이 높다 
```

```python
c=a.corr(numeric_only= True)
```

```python
c["SALARY"].sort_values() # 기본 오름차순 
```

```python
c["SALARY"].sort_values(ascending=False) # 기본 오름차순 
```
