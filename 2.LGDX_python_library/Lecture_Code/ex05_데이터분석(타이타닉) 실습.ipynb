## 3. 전처리 응용실습

### 3-1. Age

- Age 결측치를 전체 평균 나이값으로 채우기

```python
train['Age'].fillna(train['Age'].mean().astype(int))
```

```python
train['Age']
```

- Age 결측치를 성별 평균 or 중간 값으로 채우기

```python
# 성별에 따른 나이 평균값
train.groupby('Sex')['Age'].median()
```

```python
train['Age'] = train['Age'].fillna(train.groupby('Sex')['Age'].transform('median'))
```

```python
train.tail(5)
```

### 3-2. 가장 빈도가 많은 값으로 Embarked 결측치 채우기

```python
train['Embarked'].value_counts().index[0]
```

```python
train['Embarked']=train['Embarked'].fillna(train['Embarked'].value_counts().idxmax())
```

```python
train['Embarked'].isnull().sum()
```

```python
train['Age'].isnull().sum()
```

## 4. 데이터 분석

- 성별에 따른 생존자 비율 
- 성별에 따른 사망자 비율

```python
train.groupby('Sex')['Survived'].value_counts(1)*100
```

```python
train[['Sex','Survived']].groupby('Sex')['Survived'].value_counts(1)*100
```

```python
train['Sex'].value_counts()
```

- 나이대별 생존/사망 비율 확인

```python
bins=list(range(int(train['Age'].min()),int(train['Age'].max()+1),20 ))
lb = ['미성년자','청년','중년','장년']
train.insert(5, 'Cats', value = pd.cut(train['Age'], bins, labels=lb))
```

```python
pd.DataFrame((train.groupby('Cats')['Survived'].value_counts(1)*100).round(1))
```

```python
pd.DataFrame((train.groupby(['Sex','Cats'])['Survived'].value_counts(1)*100).round(1))
```
