### 요일별 사고 건 수를 시각화하기


```python
data2 = data["요일"].value_counts()
```

```python
day = data2[["월","화","수","목","금","토","일"]]
```

```python
day.values
```

```python
day.index
```

```python
#시각화 할 수 있는 라이브러리 불러오기!
import matplotlib.pyplot as plt
plt.rc('font',family="Malgun Gothic")
```

```python
#bar 그래프 그려주기 -> x,y축에 어떤 데이터가 들어갈건지!
x = day.index
y = day.values
plt.bar(x,y)
plt.show()
```

```python
#그래프 색 변경하기
plt.bar(x,y,width=0.3,edgecolor='r',color='y')
#500~620사이의 데이터를 명백히 구분할 수 있게하기
plt.ylim(500,620)
#x축 이름 설정하기
plt.xlabel("요일",labelpad=10,loc="left")
#y축 이름 설정하기
plt.ylabel("사고건수",labelpad=50,rotation=0)
#그래프의 이름 설정하기
plt.title("요일별 사고건수")
plt.show()
```

### 사고유형 중 차대차 사건 중 죽거나 다친 사람이 많은 발생지시도를 알아보고 시각화해보자!


```python
data.columns
```

```python
data['사고유형_대분류']
```

```python
#사고유형 대분류 내 각 Value값이 얼마나 있는 지 살펴보자!
data['사고유형_대분류'].value_counts()
```

```python
#사고유형 대분류 중 "차대차" 교통사고만 가져오기! -> boolean인덱싱활용하기!
pd.set_option('display.max_columns',None)
car = data[data["사고유형_대분류"] == "차대차"]
car
```

```python
car.columns
```

```python
#car 데이터에서 각 발생지시도에서 발생한 사상자수 확인해보기!
data1 = car[["사상자수","발생지시도"]]
data1
```

```python
#각 발생지시도의 행의 갯수를 출력하는 value_counts()
data1["발생지시도"].value_counts()
```

```python
#우리는 각 발생지시도에서 몇명의 사상자가 나왔는지 그 사상자의 총합이 필요하다!!
#groupby -> 같은 값을 하나로 묶어서 총계,합계 결과를 가져오자!
result = data1.groupby('발생지시도').sum()
result
```

```python
#라인그래프로 보여주기!
x = result.index #발생지시도
y = result.values #사상자수
plt.plot(x,y)
plt.show()
#이런 선그래프는 시계열 데이터에 적합하다!
```
