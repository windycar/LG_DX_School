### 네이버 뉴스데이터 수집
- 실습 프로세스
  1. 네이버 뉴스페이지 접속 -> IT/과학 카테고리
  2. 헤드라인 뉴스 접근 (주소 받아오기!)
  3. 뉴스의 제목, 내용 수집

```python
# 라이브러리 불러오기
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 페이지 정보 불러오기
url = 'https://n.news.naver.com/mnews/article/366/0001136987'
res = req.get(url)
soup = bs(res.text,'lxml')
soup
```

```python
# 뉴스 제목 추출
soup.select('h2#title_area>span')[0].text
```

```python
# 뉴스 내용 추출
soup.select_one('article#dic_area').text.strip()
```
