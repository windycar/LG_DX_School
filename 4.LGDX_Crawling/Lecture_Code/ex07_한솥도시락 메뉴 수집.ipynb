### 학습목표
- 한솥도시락의 메뉴며으 가격을 수집하여 표로 정리할 수 있다.
- 셀레니움 라이브러리를 활용하여 한솥도시락의 모든메뉴를 수집할 수 있다.

```python
# 라이브러리 불러오기 (기억을 더듬어서 작성 -> 보고 작성)
# 웹드라이버, 기준, 컴퓨터용 키보드, 쉬는시간
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
# 페이지 요청 후 열기
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(2)
```

```python
# 데이터 수집 (menu,prcie -> 요소)
menu = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
price = driver.find_elements(By.CSS_SELECTOR, 'div.item-price>strong')
```

```python
len(menu), len(price)
```

```python
# 출력 (순수한 콘텐츠만 담긴 결과 , menu_list, price_list)
# menu_list = [실행문 for]
menu_list = [i.text for i in menu]
price_list = [i.text for i in price]
```

##### 예외처리
- 키워드: try, except
- 구조  
try:  
(들여쓰기)실행문  
except:  
(들여쓰기)오류가 발생했을시 진행할 실행문
- 실행문을 실행하다가 오류가 발생했을때 except 내의 코드를 실행
- 오류 발생시 코드가 멈추지 않고 계속해서 돌아갈 수 있게끔 해줌!

```python
# 반복문 for 문을 활용하여 더보기버튼 연속해서 누르기!
# for -> 20번 -> 더보기 버튼이 20개가 존재하지 않는다면 오류발생!
# 예외처리를 통하여 모든 페이지를 오류없이 열기
try:
    for i in range(20):
        btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        btn.click()
        time.sleep(1)
except:
    print('더보기 완료!')
```

- 코드 합치기~

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 페이지 요청 후 열기
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(2)

try:
    for i in range(20):
        btn = driver.find_element(By.CSS_SELECTOR, 'a.c_05')
        btn.click()
        time.sleep(1)
except:
    print('더보기 완료!')

# 데이터 수집 (menu,prcie -> 요소)
menu = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
price = driver.find_elements(By.CSS_SELECTOR, 'div.item-price>strong')

menu_list = [i.text for i in menu]
price_list = [i.text for i in price]

# DataFrame 으로 만들기~
dic_hansot = {"메뉴명":menu_list, "가격":price_list}
df_hansot = pd.DataFrame(dic_hansot)
df_hansot.to_excel('한솥 메뉴 정리.xlsx')
```
