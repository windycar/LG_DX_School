### 학습목표
- 멜론차트 top 100 에 있는 노래제목, 가수이름 수집할 수 있다.
- 다양한 환경에 있는 태그를 활용하여 원하는 데이터를 수집할 수 있다.

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 멜론차트 페이지 정보 받아오기~
url = 'https://www.melon.com/chart/'
req.get(url)
# Response [406]: 멜론페이지에서 브라우저가 아닌 코드로 접근한것을 인지 -> 접근불가
```

```python
h = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36'}
url = 'https://www.melon.com/chart/'
res = req.get(url, headers = h) # 브라우저 정보를 포함시켜 요청
res
```

```python
soup = bs(res.text,'lxml')
soup
```

### 데이터 수집
- 노래 제목
- 가수

```python
# 1. 노래제목 수집 (요소를 title 변수에 담기)
title = soup.select('div.ellipsis.rank01>span>a')
len(title)
```

```python
# 2. 가수 수집
singer = soup.select('div.ellipsis.rank02>a')
len(singer)
```

```python
title[0].text
```

```python
singer[0].text
```

```python
singer = soup.select('div.ellipsis.rank02>span')
len(singer)
```

```python
singer[1].text
```

```python
music_list = []
for i in range(100):
    t = title[i].text
    s = singer[i].text
    music_list.append([t,s])
```

```python
music_list
```

```python
top100 = pd.DataFrame(music_list, columns=['노래제목','가수'])
top100.index = top100.index + 1
```

```python
top100.index.name = '순위'
```

```python
top100
```

```python
# DataFrame 생성시 한번에 컬럼명, 인덱스명, 인덱스번호 지정하기
top100 = pd.DataFrame(music_list, 
                      columns=['노래제목','가수'],
                     index = pd.Index(range(1, 101),name = '순위'))
top100
```
