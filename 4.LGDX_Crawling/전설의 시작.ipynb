```python
import time
import random
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, StaleElementReferenceException
from selenium import webdriver as wb
# ì»´í“¨í„°ìš© í‚¤ë³´ë“œ (Enter,Del ë“± ) ì»´í“¨í„°ì—ê²Œ í‚¤ë³´ë“œë¥¼ ì „ë‹¬í•˜ëŠ” ì—­í•  
from selenium.webdriver.common.keys import Keys 
# ì„ íƒìë¥¼ êµ¬ë¶„í• ë•Œ ì‚¬ìš©í•˜ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬ (í•„ìˆ˜)
from selenium.webdriver.common.by import By
driver = wb.Chrome()
driver.get(url)
driver.switch_to.frame('myFrame')
driver.find_elements(By.CSS_SELECTOR,'#loginBtn')[0].click()
search=driver.find_element(By.CSS_SELECTOR,'#uuid-f > div:nth-child(1) > input')
search.send_keys('20203109')
search=driver.find_element(By.CSS_SELECTOR,'#uuid-g > div:nth-child(1) > input')
search.send_keys('Jyc6863972!')
search.send_keys(Keys.ENTER)
# ì„œë²„ í­ì£¼ë¥¼ ëŒ€ë¹„í•´ ëŒ€ê¸° ì‹œê°„ì„ ë„‰ë„‰íˆ(30ì´ˆ) ì¡ìŠµë‹ˆë‹¤.
# (í•˜ì§€ë§Œ ì„œë²„ê°€ ì‘ë‹µí•˜ë©´ 0.01ì´ˆ ë§Œì— ë°”ë¡œ ì‹¤í–‰ë©ë‹ˆë‹¤.)
heavy_wait = WebDriverWait(driver, 30) 

for i in range(1, 6):
    retry_count = 0
    while retry_count < 3: # ì„œë²„ê°€ ì‘ë‹µ ì—†ì„ ì‹œ í–‰ë§ˆë‹¤ ìµœëŒ€ 3ë²ˆ ì¬ì‹œë„
        try:
            print(f"ğŸ”¥ {i+1}í–‰ ì‹œë„ ì¤‘ (ì¬ì‹œë„: {retry_count})...")
            
            # 1. íƒ€ê²Ÿ ë²„íŠ¼ì´ ë‚˜íƒ€ë‚  ë•Œê¹Œì§€ ëˆì§ˆê¸°ê²Œ ëŒ€ê¸°
            target_css = f"div.cl-text-wrapper[role='button'][aria-label='{i+1}í–‰ 1ì—´']"
            target_btn = heavy_wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, target_css)))

            # 2. í´ë¦­ ì´ë²¤íŠ¸ ì£¼ì… (ì„œë²„ ë¶€í•˜ ì‹œ JS í´ë¦­ì´ ê°€ì¥ í™•ì‹¤í•¨)
            driver.execute_script("""
                var el = arguments[0];
                ['mousedown', 'mouseup', 'click'].forEach(type => {
                    el.dispatchEvent(new MouseEvent(type, { 'bubbles': true, 'view': window }));
                });
            """, target_btn)
            
            # 3. íŒì—… í™•ì¸ ë²„íŠ¼ ëŒ€ê¸° (ì„œë²„ê°€ ëŠë¦¬ë©´ ì—¬ê¸°ì„œ ëŒ€ê¸° ë°œìƒ)
            # íŒì—…ì´ ëœ° ë•Œê¹Œì§€ ê¸°ë‹¤ë¦¼
            confirm_btn = heavy_wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div.cl-button[class*='popup-btn']")))
            driver.execute_script("arguments[0].click();", confirm_btn)
            
            print(f"âœ… {i+1}í–‰ ì„±ê³µ!")
            
            # ì„±ê³µí–ˆìœ¼ë¯€ë¡œ ë‹¤ìŒ í–‰ìœ¼ë¡œ ë„˜ì–´ê°€ê¸° ìœ„í•´ whileë¬¸ íƒˆì¶œ
            time.sleep(random.uniform(0.2, 0.5)) # ìµœì†Œí•œì˜ ì¸ê°„ë¯¸ ë”œë ˆì´
            break 

        except (TimeoutException, StaleElementReferenceException):
            retry_count += 1
            print(f"âš ï¸ {i+1}í–‰ ì„œë²„ ì‘ë‹µ ì§€ì—°... ì¬ì‹œë„ ì¤‘ ({retry_count}/3)")
            time.sleep(0.5) # ì¬ì‹œë„ ì „ ì ê¹ì˜ íœ´ì‹
            
            if retry_count == 3:
                print(f"âŒ {i+1}í–‰ ê²°êµ­ ì‹¤íŒ¨ (ì„œë²„ ë§ˆë¹„)")
```
