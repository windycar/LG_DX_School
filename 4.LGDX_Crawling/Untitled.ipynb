```python
n = 5  # í”¼ë¼ë¯¸ë“œ ì¤‘ê°„ì˜ ìµœëŒ€ ë³„ ê°œìˆ˜

# 1. ìœ„ìª½: ë³„ì´ ëŠ˜ì–´ë‚˜ëŠ” êµ¬ê°„ (1ì¸µë¶€í„° nì¸µê¹Œì§€)
for i in range(1, n + 1):
    for j in range(i):
        print('*', end='')
    print()

# 2. ì•„ë˜ìª½: ë³„ì´ ì¤„ì–´ë“œëŠ” êµ¬ê°„ (n-1ì¸µë¶€í„° 1ì¸µê¹Œì§€)
for i in range(n - 1, 0, -1):
    for j in range(i):
        print('*', end='')
    print()
```

```python
n = 5  # ìµœëŒ€ ë³„ ê°œìˆ˜

# 1. ìœ„ìª½: ë³„ì´ ëŠ˜ì–´ë‚˜ê³  ê³µë°±ì€ ì¤„ì–´ë“œëŠ” êµ¬ê°„
for i in range(1, n + 1):
    # ê³µë°± ì¶œë ¥ ë£¨í”„
    for j in range(n - i):
        print(' ', end='')
    # ë³„ ì¶œë ¥ ë£¨í”„
    for k in range(i):
        print('*', end='')
    print()

# 2. ì•„ë˜ìª½: ë³„ì´ ì¤„ì–´ë“¤ê³  ê³µë°±ì€ ëŠ˜ì–´ë‚˜ëŠ” êµ¬ê°„
for i in range(n - 1, 0, -1):
    # ê³µë°± ì¶œë ¥ ë£¨í”„
    for j in range(n - i):
        print(' ', end='')
    # ë³„ ì¶œë ¥ ë£¨í”„
    for k in range(i):
        print('*', end='')
    print()
```
