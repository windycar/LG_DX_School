```python
def preprocess_sentence_kr(w):
  w = w.strip()
  w = re.sub(r"[^0-9가-힣?.!,¿]+", " ", w) 
  w = w.strip() 
  return w
 
```

```python
url='https://www.pokemonkorea.co.kr/pokedex#pokedex_1'
```

```python
driver = wb.Chrome()
driver.get(url)
```

```python
body=driver.find_element(By.TAG_NAME,'body')
body.send_keys(Keys.END)
```

```python
img_list=[]
aTag=driver.find_elements(By.CSS_SELECTOR,'.col-lg-2.col-6>a>div>div>img')
```

```python
for i in aTag:
     img_list.append(i.get_attribute('src'))
img_list  
```

- 이미지 파일 로컬에 저장하기 

```python
#urlretrieve(url,'경로.확장자 ')
urlretrieve(img_list[0],r'C:\Users\jyc68\LG_DX_school_code\data\image_poke\포켓몬.png')
```

```python
name_list=[]
name=driver.find_elements(By.CSS_SELECTOR,'h3')
name[0].text
for i in name:
     name_list.append(preprocess_sentence_kr(i.text))
```

```python
for i in range(len(img_list)):
     urlretrieve(img_list[i],f'C:\\Users\\jyc68\\LG_DX_school_code\\data\\image_poke\\{name_list[i]}.png')
```
