```python
url='https://news.naver.com/section/105'
new=req.get(url)
news= bs(new.text,'lxml')
#풀이 
```

```python
new_url=news.select('a.sa_text_title._NLOG_IMPRESSION')
```

```python
#get('속성명')
new_url[0].get('href')
```

```python
len(new_url)
```

```python
link_list=[]
for i in range(len(new_url)):
    link_list.append(new_url[i].get('href'))
    print(link_list[i])
```

```python
url='https://news.naver.com/section/105'
new=req.get(url)
news= bs(new.text,'lxml')
#풀이 
```

```python
data_list=[]
for i in range(len(link_list)):
    url=link_list[i]
    new=req.get(url)
    news= bs(new.text,'lxml')

    title=news.select("h2#title_area>span")[0].text
    content=news.select_one('article#dic_area').text.strip()
    data_list.append([title,content, url])
    






    
```

```python
len(data_list)
```

```python
import pandas as pd
```

```python
news_df=pd.DataFrame(data_list,columns=['뉴스제목','내용','링크'])
```
