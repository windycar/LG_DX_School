```python
import requests as req 
from bs4 import BeautifulSoup as bs
```

```python
url='https://n.news.naver.com/mnews/article/366/0001136987'
```

```python
#뉴스 제목 추출
new=req.get(url)
```

```python
news= bs(new.text,'lxml')
```

```python
title=news.select_one("h2.media_end_head_headline>span")
#풀이 
```

```python
print(title.text)
```
