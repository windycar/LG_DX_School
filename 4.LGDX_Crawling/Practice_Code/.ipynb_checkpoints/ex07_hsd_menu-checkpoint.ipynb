### 학습목표 
- 한솥도시락의 메뉴명의 가격을 수집하여 표로 정리할수있다
- 셀레니움 라이브러리를 활용하여 한솥의 모든메뉴를 수집할수있다 

```python
from tqdm import tqdm
import pandas as pd
import requests as req 
from bs4 import BeautifulSoup as bs
import re
from kiwipiepy import Kiwi
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
# 컴퓨터용 브라우저 -> 데이터 요청 및 응답 
from selenium import webdriver as wb
# 컴퓨터용 키보드 (Enter,Del 등 ) 컴퓨터에게 키보드를 전달하는 역할 
from selenium.webdriver.common.keys import Keys 
# 선택자를 구분할때 사용하는 라이브러리 (필수)
from selenium.webdriver.common.by import By
#쉬는 시간을 주는 라이브러리 
import time
```

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(1)
menu_list=[]
price_list=[]
a=driver.find_elements(By.CSS_SELECTOR,'.item-text h4')
b=driver.find_elements(By.CSS_SELECTOR,'.item-text div>strong')
for i in range(len(a)):
    menu_list.append(a[i].text)
    price_list.append(b[i].text)
driver.quit()    
print(menu_list)
print(price_list)
```

```python
data = pd.DataFrame({
    '메뉴이름': menu_list,
    '가격': price_list
})
```

##### 예외처리
- 키워드: try, except
- 구조  
try:  
(들여쓰기)실행문  
except:  
(들여쓰기)오류가 발생했을시 진행할 실행문
- 실행문을 실행하다가 오류가 발생했을때 except 내의 코드를 실행
- 오류 발생시 코드가 멈추지 않고 계속해서 돌아갈 수 있게끔 해줌!

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(1)

try:
    for i in range(20):
        driver.find_element(By.CSS_SELECTOR,'a.c_05').click()
        time.sleep(1)
except:
    menu_list=[]
    price_list=[]
    a=driver.find_elements(By.CSS_SELECTOR,'.item-text h4')
    b=driver.find_elements(By.CSS_SELECTOR,'.item-text div>strong')
    for i in tqdm(range(len(a))):
        menu_list.append(a[i].text)
        price_list.append(b[i].text)
    driver.quit()       
```

```python
print(menu_list)
print(price_list)
```

```python
data = pd.DataFrame({
    '메뉴이름': menu_list,
    '가격': price_list
})
```
