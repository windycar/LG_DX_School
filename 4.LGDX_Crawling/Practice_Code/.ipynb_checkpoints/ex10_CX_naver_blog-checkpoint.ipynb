```python
from tqdm import tqdm
import pandas as pd
import requests as req 
from bs4 import BeautifulSoup as bs
import re
from kiwipiepy import Kiwi
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
# 컴퓨터용 브라우저 -> 데이터 요청 및 응답 
from selenium import webdriver as wb
# 컴퓨터용 키보드 (Enter,Del 등 ) 컴퓨터에게 키보드를 전달하는 역할 
from selenium.webdriver.common.keys import Keys 
# 선택자를 구분할때 사용하는 라이브러리 (필수)
from selenium.webdriver.common.by import By
#쉬는 시간을 주는 라이브러리 
import time
#mask 이미지를 만들기 위한 도구
from PIL import Image
import numpy as np
# 글(문자열등) 데이터가 url에 들어갓을때 띄어쓰기가 있으면 오류 발생
# url 글이 들어갔을떄 안전하게 사용하수있는 혀애로 변경해주는 기능 
from urllib.parse import quote
```

```python
def preprocess_sentence_kr(w):
  w = w.strip()
  w = re.sub(r"[^0-9가-힣?.!,¿]+", " ", w) 
  w = w.strip() 
  return w
 
```

```python
keyword=quote('음식물 처리기 사용 후기')
# date_option 패턴: 0(전체), 1h(1시간),1d(1일),1w(1주),1m(1개월),3m(3개월),6m(6개월), 1y(1년)
data_option='6m'
url=f'https://search.naver.com/search.naver?ssc=tab.blog.all&query={keyword}&sm=tab_opt&nso=so%3Ar%2Cp%3A{data_option}'
```

```python
url
```

```python
driver = wb.Chrome()
driver.get(url)
```

```python
body=driver.find_element(By.TAG_NAME,'body')
body.send_keys(Keys.END)
for i in range(10):
    body=driver.find_element(By.TAG_NAME,'body')
    body.send_keys(Keys.END)
    time.sleep(0.3)
```

```python
href_list=[]
aTag=driver.find_elements(By.CSS_SELECTOR,'.sds-comps-base-layout.sds-comps-full-layout.EXA2vv5sZMziIkX_QWv9>a')


```

```python

for i in aTag:
     href_list.append(i.get_attribute('href'))
href_list  
```

```python
 len(href_list )
```

```python
driver.get(href_list[0])
```

```python
driver.switch_to.frame('mainFrame')
```

```python
driver.find_elements(By.CSS_SELECTOR,'.se-main-container')[0].text
```

```python

f=open('../../data/naver_blog.txt','w',encoding='utf-8')
for i in tqdm(range(len(href_list))):
    driver.get(href_list[i])
    time.sleep(0.3)
    driver.switch_to.frame('mainFrame')
    a=driver.find_elements(By.CSS_SELECTOR,'.se-main-container')
    for j in a:
        content=preprocess_sentence_kr(j.text)
        f.write(content)
f.close()        
driver.quit()    
```

```python
f=open('../../data/naver_blog.txt','r',encoding='utf-8')
review=f.readline()
f.close()


```

```python
kiwi=Kiwi()
#토큰화
token=kiwi.tokenize(review)
```

```python

# 일반 명사만 추출 (NNG)
# token 의 리스트를 돌면서 tag가 NNG 인 form을 nng_list에 답으세요
nn_list=[]
for i in tqdm(token):
    if(i.tag=='NNG' or i.tag=='NNP'):
        nn_list.append(i.form)
```

```python
counter=Counter(nn_list)
top_100=counter.most_common(100)
top_100
```

```python
img=Image.open('../../data/wlwldyd.jpg')
mak_img=np.array(img)
```

```python
wc = WordCloud(
    font_path='C:/Windows/Fonts/malgunbd.ttf',
    background_color='white'  ,  
    mask=mak_img,
    colormap='winter'
).generate_from_frequencies(dict(top_100))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
