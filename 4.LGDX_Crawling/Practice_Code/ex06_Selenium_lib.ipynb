### 실습 
- 네이버 메인 페이지를 열어 검색어 결과 수집하기 

##### 코드작성 순서
1. 브라우저 열기
2. 네이버 페이지로 이동하기
3. 검색창 클릭하기
4. 검색어 입력학
5. 엔터 

```python
#1.브라우저 열기
driver = wb.Chrome()
```

```python
# 2.네이버 페이지 요청 
# dirver -> 크롬창이 열린상태에서 명령 
driver.get('https://www.naver.com/')
```

```python
# 3. 검색어 입력하기 (keyword: 강아지 )
# 검색창위치를 드라이버에게 공유 query
search=driver.find_element(By.ID,'query')
#4.검색어 입력
search.send_keys('강아지')
```

```python
#5.엔터
search.send_keys(Keys.ENTER)
```

```python
# 페이지에 있느 강아지ㅡ이 종류 수집
a=driver.find_elements(By.CSS_SELECTOR,'a._text')[0].text
```

```python
# 페이지에 있느 강아지ㅡ이 종류 수집 출력 

a=driver.find_elements(By.CSS_SELECTOR,'a._text')
for i in a:
    print(i.text)
```
