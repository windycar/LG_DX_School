- 순서
    1.url 주소를 통해 page 정보 받아오기
    2.응답받은 무서 html 문서화
    3.뉴스제목 ,주소수집후 리스트에저장

```python
import pandas as pd
import requests as req 
from bs4 import BeautifulSoup as bs
```

```python
url='https://search.zdnet.co.kr/?kwd=%EC%97%94%EB%B9%84%EB%94%94%EC%95%84&area=&pageno=1&term='
new=req.get(url)
news= bs(new.text,'lxml')
```

```python
ab=news.select('section.news_box div.assetText>a')
len(ab)
```

```python
ab[0].get('href')
```

```python
link_list=[]
for i in range(len(ab)):
    link_list.append(ab[i].get('href'))
    print(link_list[i])
```

```python
url1='https://zdnet.co.kr/view/?no=20260120100934'
new1=req.get(url1)
news1= bs(new1.text,'lxml')
```

```python
news1.select('h1')
```

```python
news1.select_one('#content-20260120100934').text.strip()
```

```python
data_list=[]
for i in range(len(link_list)):
    url=link_list[i]
    new=req.get(url)
    news= bs(new.text,'lxml')

    title=news.select("h1")[0].text
   
    data_list.append([title, url])
    




```

```python
news_df=pd.DataFrame(data_list,columns=['뉴스제목','링크'])
```

### 10페이지 뉴스데이터 수집

```python
data_list=[]
for i in tqdm(range(10)):
    url=f'https://search.zdnet.co.kr/?kwd=%EC%97%94%EB%B9%84%EB%94%94%EC%95%84&area=&pageno={i+1}&term='
    new=req.get(url)
    news= bs(new.text,'lxml')
    ab=news.select('section.news_box div.assetText>a')
    link_list=[]
    for j in range(len(ab)):
        link_list.append(ab[j].get('href'))
    
    for j in range(len(link_list)):
        url=link_list[j]
        new=req.get(url)
        news= bs(new.text,'lxml')

        title=news.select("h1")[0].text
   
        data_list.append([title, url])
news_df=pd.DataFrame(data_list,columns=['뉴스제목','링크'])     
```
