```python
from tqdm import tqdm
import pandas as pd
import requests as req 
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.melon.com/chart/'
print(req.get(url))
new=req.get(url)
#<Response [406]> 멜론에서 브라우저가 아닌 코드로 접근한것을 인지 -> 접근 불가 
chart= bs(new.text,'lxml')
```

```python
h={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'}
```

```python
url='https://www.melon.com/chart/'

new=req.get(url, headers =h)
print(new)
#<Response [406]> 멜론에서 브라우저가 아닌 코드로 접근한것을 인지 -> 접근 불가 
chart= bs(new.text,'lxml')

```

```python
a=chart.select(".ellipsis.rank01>span")
b=chart.select(".ellipsis.rank02>span")
len(b)
```

```python
total=[]
for i in range(len(a)):
    titletmp=(a[i].text.replace("\n",''))
    singertmp=(b[i].text)
    total.append([titletmp,singertmp])
```

```python
chart1=pd.DataFrame(total,columns=['노래제목','가수'],index=pd.Index(range(1,101),name='순위'))
```
