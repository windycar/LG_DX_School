#### 문제7

```python
score_list=[['김미희',89,78,99],['서대희',67,88,77],['강예진',100,55,96]]
def grade(mean):
    if mean>=90:
        return 'A'
    elif mean>=80:
        return 'B'
    elif mean>=70:
        return 'C'
    elif mean>=60:
        return 'D'
    else:
        return 'F'
print('='*50)
print("이름\tPython\tJava\tDB\t평균\t학점")
print('='*50)
for i in range(len(score_list)): 
    tmp_mean=int((score_list[i][1]+score_list[i][2]+score_list[i][3])/3)
    for j in range(len(score_list[i])):
        print(f"{score_list[i][j]}",end='\t')
    print(f"{tmp_mean}\t{grade(tmp_mean)}")
print('='*50)
print("평균",end='\t')

for i in range(1,4):
    tmp=0
    for j in range(len(score_list)):
        tmp+=score_list[j][i]
    print(int(tmp/len(score_list)),end='\t')
```

```python
score_list=[['김미희',89,78,99],['서대희',67,88,77],['강예진',100,55,96],['차현석',75,100,60]]
print('='*50)
print("이름\tPython\tJava\tDB\t평균\t학점")
print('='*50)
for i in range(len(score_list)): 
    tmp_mean=int((score_list[i][1]+score_list[i][2]+score_list[i][3])/3)
    for j in range(len(score_list[i])):
        print(f"{score_list[i][j]}",end='\t')
    print(f"{tmp_mean}\t{grade(tmp_mean)}")
print('='*50)
print("평균",end='\t')

for i in range(1,4):
    tmp=0
    for j in range(len(score_list)):
        tmp+=score_list[j][i]
    print(int(tmp/len(score_list)),end='\t')
```

#### 문제9

```python
#9-1
import pandas as pd
team = pd.DataFrame([['홍길동',70,80,88],
                    ['성춘향',80,90,78],
                    ['김사또',90,60,90]],
                   columns = ['이름','국어','산수','과학'],)
team
```

```python
#9-2
team['평균']=team[['국어','산수','과학']].mean(axis=1)
team
```

```python
team['평균']=team['평균'].round(1)
team.loc['종합']=team[['국어','산수','과학','평균']].mean().round(1)
team.loc['종합']=team.loc['종합'].fillna('-')
```
