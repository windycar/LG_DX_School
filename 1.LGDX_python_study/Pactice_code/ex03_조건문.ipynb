```python
score=int(input("점수를 입력하시오 "))
if score>=90 and score<=100:
    print(f"{score}점은 A학점입니다")
elif score>=80:
    print(f"{score}점은 B학점입니다")
elif score>=70:
    print(f"{score}점은 C학점입니다")
elif score>=60:
    print(f"{score}점은 D학점입니다")
else :
    print(f"{score}점은 F학점입니다")
```

```python
print("----[SM Coffee]----")
money=int(input("금액을 입력하시오 "))
print("커피를 고르세요 ")
coffee=int(input("1.에스프레소 600원 2. 아메리카노 1000원 3.카페라떼 1300원>> "))
if coffee==1 and money>600 :
    charge= money-600
    charge_1000=charge//1000
    charge_500=(charge%1000)//500
    charge_100=(charge%500)//100
    print(f"거스름돈 {charge}원\n천원: {charge_1000}장\n오백원: {charge_500}개\n백원: {charge_100}개")
    
elif coffee==2 and money>1000 :
    charge= money-1000
    charge_1000=charge//1000
    charge_500=(charge%1000)//500
    charge_100=(charge%500)//100
    print(f"거스름돈 {charge}원\n천원: {charge_1000}장\n오백원: {charge_500}개\n백원: {charge_100}개")
elif coffee==2 and money>1300 :
    charge= money-1300
    charge_1000=charge//1000
    charge_500=(charge%1000)//500
    charge_100=(charge%500)//100
    print(f"거스름돈 {charge}원\n천원: {charge_1000}장\n오백원: {charge_500}개\n백원: {charge_100}개")
else :
    print("돈이 없어용")


  
```
