### 비교 연산자
 - 크거나 같은지 : >,>=,<,<=
 - 같거나 다른지 : ==,!=
 - 결과값 :True(1) or False(0)

```python
a=3
b=7
print(a>b)
print(a<b)
print(a==b)
print(a!=b)
```

### 멤버연산자 
 - in : 찾고자 하는값이 포함되어있으면 True
 - not in : 찾고자 하는값이 포함되어 있지않으면 True

```python
str1="오늘 점심 뭐먹어 ?"
"?" in str1
```

```python
"?" not in str1
```

```python
"!" not in str1
```

```python
# 리스트 : [ ]

"멜론" in ['딸기','수박',"참외"]
"멜론" not in ['딸기','수박',"참외"]
```

### 삼항연산자 
 - a if 논리 else b 
 

```python
num=int(input('시험성적 입력 > ') )

if num>=60:
    print(f"{num}점 60점 이상이라 합격입니다 ")
else:
    print(f"{num}점 60점 미만이라 불합격입니다 ")
```

```python
a=int(input('정수 입력 > ') )
b=int(input('정수 입력 > ') )
if a>b:
    print("a : ",a)
else :
    print("b : ",b)
```

```python
ball=int(input("농구공 수 입력 >>"))
if(ball%5!=0):
    result=ball//5+1
else:
    result=ball//5

print(f"필요한 상자의 수 : {result}박스 ")
```

```python
worktime=int(input('근무시간 입력>>'))
if worktime>9:
    money=8*10000+(worktime-8)*10000*1.5
else :
    money=worktime*10000
money=int(money)    
print(f"총지급액 : {money}원 입니다")
```

```python
DB=int(input('DB>>'))
cal=int(input('전자계산기>>'))
os=int(input('운영체제>>'))
dt=int(input('데이터통신>>'))
sw=int(input('소프트웨어공학>>'))

if(DB<8 or cal<8 or os<8 or dt<8  or sw<8 or (DB+cal+os+dt+sw)<60):
    print("불합격입니다")
else:
    print("합격입니다")

```
