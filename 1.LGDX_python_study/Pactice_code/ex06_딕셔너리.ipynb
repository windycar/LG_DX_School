### 딕셔너리
- key : value 한쌍으로 이루어진 자료 구조
- 순서가 없는 형태 (인덱스 번호가 없다)
- {} 로 생성
- key 값은 변하지 않는 유일한 값 -> 중복 사용시 나중에 사용값만 적용 

```python
student={'이름':'김소은',"나이":"21","전화번호":"01012345678"}
student
```

```python
print(type(student))
```

```python
#밸류 추출 가능 지지용
student['이름']
```

- value 값 변경가능 
  - 추가 수정 삭제 

```python
student["성별"]= "여자"
```

```python
student['나이']=24
```

```python
student
```

```python
del student['나이']
```

```python
student
```

```python
student['취미']=['코딩','넷플릭스 시청','음악감상']
student
```

```python
student.get('취미')
```

```python
student.keys()
```

```python
print(student.keys())
```

```python
student.values()
```

```python
for key in student.keys():
    print(key)
```

```python
for values in student.values():
    print(values)
```

```python
student.items()
```

```python
for data in student.items():
    for txt in data:
        print(txt)
```

```python
for key,value in student.items():
   
        print(key,value)
```
