```python
# print("어서오세요 용철 셰프 주방입니다")
menu=(["Burger", "Pizza", "Pasta", "Salad", "Sushi"])
bucket=[]
while True:
    print("[1]메뉴선택,[2]주문삭제,[3]주문요청(종료)")
    case2=int(input("선택하세요"))
    if case2==1:
        for i in range(len(menu)):
              print(menu[i])
        menu_sir=int(input("선택할 메뉴 개수를 입력하세요 "))
        for i in range(menu_sir):
            a=input("메뉴를 입력하세요")
            bucket.append(a)
        print("장바구니 : ",bucket)
    elif case2==2:
        menu_del=input("삭제할 메뉴를 입력하세요 ")
        for i in range (len(bucket)):
            if menu_del==bucket[i]:
                del bucket[i]
                break    
        print("장바구니 : ",bucket)        
    elif case2==3:
        print("[3]주문요청(종료)")
        off=input("q를 누르시면 종료 됩니다 ")
        if off=='q':
           break;
        else:
           print("잘못골랐어요 '다시' ")
    else:
        print("잘못골랐어요 '다시' ")
print("최종메뉴는 ",end='')
for i in range(0,len(bucket),1):
    print(f"{bucket[i]}",end=',')
print(" 입니다")    
```

```python
print('='*50)
studentScore=[]
list2=[['이름', 'JAVA', 'Web', 'Grade']]
while True:
    menu=int(input("[1]시험점수 입력 [2]성적확인 [3]종료 >>"))
    if menu==1:
        studentScore=[]
        tmp=input("이름 입력 : ")
        studentScore.append(tmp)
        tmp=int(input("자바 점수입력 : "))
        studentScore.append(tmp)
        tmp=int(input("web 점수입력 : "))
        studentScore.append(tmp)
        tmp=(studentScore[1]+studentScore[2])/2
        if tmp>=90:
            grade='A'
        elif tmp>=80:
            
            grade='B'
        elif tmp>=70:
            grade='C'
        elif tmp>=60:
            grade='D'
        else :
            grade='F'
        studentScore.append(grade)
        list2.append(studentScore)
    elif menu==2:
        
        for i ,j ,k, l in list2:
             print(f"{i}\t{j}\t{k}\t{l}")
    elif menu==3:
        break;
    else: 
        print ("잘못눌렀어요 다시")
    print('-'*50)    
print('='*50)        
```
