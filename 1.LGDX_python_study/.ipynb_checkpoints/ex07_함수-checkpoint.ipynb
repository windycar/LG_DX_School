### 함수
- 하나의 특별한 목적의 작업을 수행하기위해 독립적으로 코드의 집합
- 함수 사용 이유
   - 반복적 코딩 회피
   - 모듈화 가능으로 가독성 증가
   - 프로그램에 문제발생 및 기능 변경 유지보수 쉬움

```python
def big(a,b):
    if a>b:
        return a
    else:
        return b        
```

```python
print(big(300000,5))
```

```python
def cal(a,b,c):
    if c=='+':
        return a+b
    elif c=='-':
        return a-b
    else :
        return '잘못치셨습니다 아무래도'
x=int(input("수 1>>>>"))
y=int(input("수 2>>>>"))        
z=input("+ - 기호 >>>>")
print(f"{x} {z} {y} = {cal(x,y,z)}" )
```

```python
def close_ten(a,b):
    
    if abs(a-10)<abs(b-10):
        return b
    elif abs(a-10)>abs(b-10):
        return a
    elif abs(a-10)==abs(b-10):
        return 0
x=int(input("수 1>>>>"))
y=int(input("수 2>>>>"))        

print(f"{x} {y} 둘 중 10의 가까운수는  {close_ten(x,y)}" )
```

```python
def rangelist(start,end):
    lst=[]
    for i in range(start,end+1):
        lst.append(i)
    return lst

x=int(input("수 1>>>>"))
y=int(input("수 2>>>>"))  
x_y_list=rangelist(x,y)
print(x_y_list)
```

```python
def cal(a,b,c):
    """"함수 설명 기호 입력 후 사칙연산 계산 """
    if c=='+':
        return a+b
    elif c=='-':
        return a-b
    elif c=='/':
        return a/b
    elif c=='//':
        return a//b
    elif c=='%':
        return a%b
    elif c=='**':
        return a**b    
    else :
        return '잘못치셨습니다 아무래도'
x=int(input("수 1>>>>"))
y=int(input("수 2>>>>"))        
z=input("연산 기호 >>>>")
print(f"{x} {z} {y} = {cal(x,y,z)}" )
```

```python
def power_num(num,power=2):
    return num**power
```

```python
print(power_num(5))
```

```python
print(power_num(5,3))
```

```python
print(power_num(782,3))
```

```python
def all_sum(*args):
    result=0
    for i in args:
        result+=i
    return result
        
```

```python
all_sum(1,2)
```

```python
all_sum(1,2,4)
```

```python
all_sum(1,2,4,8,4654,787,2,594,216458,1654654,1654654,4787,987,1321)
```
