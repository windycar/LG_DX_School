###### 인덱싱과 슬라이싱
- 인덱싱 : 무엇인가를 가리킨다는 의미 ,인덱스를 활용해 문자를 가져오기 
- 슬라이싱 : 무엇인가를 잘라낸다라는 의미 , 인덱스를 활용해 원하는 범위 추출

```python
s= "My name is MH"
print(s[12])
print(s[8])
print(s[-2])
```

```python
print(s[8:10])
print(s[3:7])
```

```python
day="2025년 12월 31일의 날씨는 흐림입니다."
print("날짜는 ", day[0:14])
print("날씨는 ",day[-6:-4] )
print("날씨는 ",day[-6:] )
print(day[:] )
```

```python
y="2025"
m="12"
d="31"
print(y,"년",m,"월",d,"일")

```

###### 문자열 포매팅 
-문자열에 특정한 데이터를 포함시키고자 할때 사용 

 1.% 기호 포매팅
 
 2.format함수
 
 3.f문자열 포메팅
 
 

1.% 기호 포매팅

```python
y="2025"
m=12
d="31"
print("오늘은 %s년 입니다."%y)
print("오늘은 %s년 %d월 입니다."%(y,m))
print("오늘은 %s년 %d월 %s일입니다."%(y,m,d))
```

2.fomat 함수 활용

```python
print("오늘은 {}년 {}월 {}일입니다.".format(y,m,d))
```

3.f문자열 포매팅

```python
print(f"오늘은 {y}년 {m}월 {d}일입니다.")
```

```python
s=f"오늘은 {y}년 {m}월 {d}일입니다."
print(len(s))
```
